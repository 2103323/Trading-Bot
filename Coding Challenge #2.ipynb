```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
start = "2014-10-01"
end = "2021-05-31"
```

```python
symbol = ["BTC-USD","MSFT"]
```

```python
df = yf.download(symbol,start,end)
```

**MSFT**:Microsoft Corp <br>
**BTC-USD**:Bitcoin in USD

```python
df
```

```python
stock = df.swaplevel(axis = "columns").sort_index(axis = "columns")
```

```python
stock=df.Close.copy()
msft = stock["MSFT"].dropna().to_frame().copy()
msft = pd.DataFrame(msft)
btc = stock["BTC-USD"].dropna().to_frame().copy()
btc = pd.DataFrame(btc)
```

```python
btc
```

```python
msft
```

```python
msft.rename(columns = {"MSFT":"Price1"},inplace = True)
msft
```

```python
btc.rename(columns = {"BTC-USD":"Price1"},inplace = True)
btc
```

```python
msft["ret"] = msft.Price1.diff(periods = 1)
msft
```

```python
btc["ret"] = btc.Price1.diff(periods = 1)
btc
```

```python
mu1 = btc.ret.mean()
mu2 = msft.ret.mean()
```

```python
mu1
```

```python
mu2
```

```python
sigma1 = btc.ret.std()
sigma2 = msft.ret.std()
```

```python
sigma1
```

```python
sigma2
```

```python
np.sqrt(btc.ret.var())
```

```python
np.sqrt(msft.ret.var())
```

Here mu1 > mu2 hence average return of BTC-USD is greater than microsoft therefore **Higher Reward** for BTC-USD but simultaneously sigma1 > sigma2 which tell us that there is high fluctation in return.Hence **Higher Risk**

So,**Higher Risk -> Higher Reward**
